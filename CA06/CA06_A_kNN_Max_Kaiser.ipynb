# **kNN - Finding k Neighbors - Recommendation Engine**

Finding ‘k’ neighbors from sample movie data set for the movie "The Post".

What are 5 movies similar to the movie "The Post"? 


Constructing a NearestNeighbors class from an array representing our data set and ask for the 5 closest points.

```python
# Movie 'The Post'

IMDB Rating = 7.2, Biography = Yes, Drama = Yes, Thriller = No, Comedy = No, Crime = No,
Mystery = No, History = Yes

--> [7.2,1,1,0,0,0,0,1]

#same order (binary 1/0)

```

```python
# Define X an y 

samples = df.drop(['Label','Movie ID','Movie Name'], axis= 1)

X = df[['IMDB Rating','Biography','Drama','Thriller','Comedy','Crime','Mystery','History']]

# or use df.drop(['Label','Movie ID','Movie Name'], axis= 1)

y = df[['Movie Name']] #response variable 

neigh = NearestNeighbors(n_neighbors=5,algorithm='brute') #default /k-value /small dataset --> brute

# brute because we have a small dataset 

#fitting the model 
neigh.fit(X,y)

#NearestNeighbors(n_neighbors=5)

movie_index = neigh.kneighbors([[7.2,1,1,0,0,0,0,1]],return_distance=False)


# loop 
for rec_movie in movie_index: 
    print(df.iloc[rec_movie]['Movie Name'])

#Returns the elements of samples (indexes start at 0) --> manual 

#df.iloc[[28,27,29,16,9]]

```
